So in this task you are to examine kernels for SVM algorithm applied to rather simple artificial datasets.

```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
import numpy as np
```

Let's generate our dataset and take a look on it.

```python
moons_points, moons_labels = make_moons(n_samples=1000, noise=0.2, random_state=42)
plt.scatter(moons_points[:, 0], moons_points[:, 1], c=moons_labels)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

from sklearn.model_selection import train_test_split

from sklearn.metrics import f1_score, accuracy_score

from sklearn.preprocessing import PolynomialFeatures

from mlxtend.plotting import plot_decision_regions
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(moons_points, moons_labels, test_size=0.3)

print (X_train.shape)
print (Y_train.shape)
print (X_test .shape)
print (Y_test .shape)
```

```python
lin_clf = LogisticRegression(penalty='l2', C=0.8) # add some params
svm_clf = SVC(kernel='linear') # here too

### YOUR CODE HERE

lin_clf.fit (X_train, Y_train)
svm_clf.fit (X_train, Y_train)

print ('linear_accuracy = ', lin_clf.score(X_test, Y_test))
print ('svm   _accuracy = ', lin_clf.score(X_test, Y_test))
```

```python
plot_decision_regions (X=X_test, y=Y_test, clf=lin_clf)
```

```python
plot_decision_regions (X=X_test, y=Y_test, clf=svm_clf)
```

## 1.2 Kernel trick

Now use different kernels (`poly`, `rbf`, `sigmoid`) on SVC to get better results. Play `degree` parameter and others.

For each kernel estimate optimal params, plot decision regions, calculate metric you've chosen eariler.

Write couple of sentences on:

```python
### YOUR CODE HERE
svm_clf_poly    = SVC(kernel='poly'   , degree=3, coef0=1) # 3 is optimal here with 2 and 4 results are worse
svm_clf_rbf     = SVC(kernel='rbf'    )
svm_clf_sigmoid = SVC(kernel='sigmoid')

svm_clf_poly   .fit (X_train, Y_train)
svm_clf_rbf    .fit (X_train, Y_train)
svm_clf_sigmoid.fit (X_train, Y_train)

print ('svm poly    accuracy = ', svm_clf_poly   .score(X_test, Y_test))
print ('svm rbf     accuracy = ', svm_clf_rbf    .score(X_test, Y_test))
print ('svm sigmoid accuracy = ', svm_clf_sigmoid.score(X_test, Y_test))
```

## 1.3 Simpler solution (of a kind)
What is we could use Logisitc Regression to successfully solve this task?

Feature generation is a thing to help here. Different techniques of feature generation are used in real life, couple of them will be covered in additional lectures.

In particular case simple `PolynomialFeatures` ([link](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.PolynomialFeatures.html)) are able to save the day.

Generate the set of new features, train LR on it, plot decision regions, calculate metric.

* Comare SVM's results with this solution (quality, borders type)
* What degree of PolynomialFeatures have you used? Compare with same SVM kernel parameter.

```python
### YOUR CODE HERE
poly = PolynomialFeatures(degree=4)

poly.fit (X_train)
X_train_poly = poly.transform (X_train)
X_test_poly  = poly.transform (X_test)

lin_clf_poly = LogisticRegression ()
lin_clf_poly.fit (X_train_poly, Y_train)

print ('polynomial features accuracy = ', lin_clf_poly.score(X_test_poly, Y_test))

```

```python
class complex_clf ():
    
    def __init__ (self, tranformator, classificator):
        self._trans = tranformator;
        self._clf = classificator
            
    def predict (self, X):
        return self._clf.predict(self._trans.transform(X))
```

```python
plot_decision_regions (X=X_test, y=Y_test, clf=complex_clf (poly, lin_clf_poly))
```

## 1.4 Harder problem

Let's make this task a bit more challenging via upgrading dataset:

```python
from sklearn.datasets import make_circles

circles_points, circles_labels = make_circles(n_samples=1000, noise=0.06, random_state=42)

plt.figure(figsize=(5, 5))
plt.scatter(circles_points[:, 0], circles_points[:, 1], c=circles_labels)
```

And even more:

```python
points = np.vstack((circles_points*2.5 + 0.5, moons_points))
labels = np.hstack((circles_labels, moons_labels + 2)) # + 2 to distinct moons classes

plt.figure(figsize=(5, 5))
plt.scatter(points[:, 0], points[:, 1], c=labels)
```

Now do your best using all the approaches above!

Tune LR with generated features, SVM with appropriate kernel of your choice. You may add some of your loved models to demonstrate their (and your) strength. Again plot decision regions, calculate metric.

Justify the results in a few phrases.

```python
### YOUR CODE HERE
X_train, X_test, Y_train, Y_test = train_test_split(points, labels, test_size=0.3)
```

```python
### YOUR CODE HERE

poly_tfr = PolynomialFeatures(degree=6)

poly_tfr.fit (X_train)
X_train_poly = poly_tfr.transform (X_train)
X_test_poly  = poly_tfr.transform (X_test)

# svm_clf_poly = LogisticRegression ()
svm_clf_poly = SVC(kernel='linear') # other perform worse
svm_clf_poly.fit (X_train_poly, Y_train)

print ('polynomial features accuracy = ', svm_clf_poly.score(X_test_poly, Y_test))
```

```python
plot_decision_regions (X=X_test, y=Y_test, clf=complex_clf (poly_tfr, svm_clf_poly))
```
